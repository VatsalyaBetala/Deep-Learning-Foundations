```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
```

```python
# Define 4 by 4 original patch
orig_4_4 = np.array([[1, 3, 5,3 ], [6,2,0,8], [4,6,1,4], [2,8,0,3]])
print(orig_4_4)
```

```python
def subsample(x_in):
# TO DO -- write the subsampling routine
  # Replace this line

  out_height = int(np.floor(x_in.shape[0]/2))
  out_width = int(np.floor(x_in.shape[1]/2))

  x_out = np.zeros((out_height, out_width))

  for i in range(out_height):
    for j in range(out_width) :
      x_out[i,j] = x_in[2*i, 2*j]

  return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Subsampled:")
print(subsample(orig_4_4))
```

```python
# Now let's try max-pooling
def maxpool(x_in):
  # TO DO -- write the maxpool routine
  # Replace this line
  out_height = int(np.floor(x_in.shape[0]/2))
  out_width = int(np.floor(x_in.shape[1]/2))
  x_out = np.zeros((out_height, out_width))

  for i in range(out_height):
    for j in range(out_width):
      x_out[i,j] = np.max(x_in[2*i:2*i+2, 2*j:2*j+2])
  return x_out
```

```python
# Finally, let's try mean pooling
def meanpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  for i in range(x_out.shape[0]):
    for j in range(x_out.shape[1]):
      x_out[i,j] = np.mean(x_in[2*i : 2*i+2, 2*j:2*j+2])
  return x_out
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[6, 8], [8,4]])
print(orig_2_2)
```

```python
# Let's first use the duplication method
def duplicate(x_in):

  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2))
  # TO DO -- write the duplication routine
  # Replace this line
  for i in range(x_in.shape[0]):
    for j in range(x_in.shape[1]):
      x_out[i*2:i*2+2, j*2:j*2+2] = x_in[i,j]
  return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Duplicated:")
print(duplicate(orig_2_2))
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in, x_high_res):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TO DO -- write the max unpooling routine
  # Replace this line
  for i in range(x_in.shape[0]):
    for j in range(x_in.shape[1]):
      block = x_high_res[2*i:2*i+2, 2*j:2*j+2]
      max_index = np.unravel_index(np.argmax(block, axis=None), block.shape)
      x_out[2*i+max_index[0], 2*j+max_index[1]] = x_in[i,j]

  return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
import numpy as np

def bilinear(x_in):
    # Initialize the output matrix with zeros; it will have twice the dimensions of the input
    out_height, out_width = x_in.shape[0] * 2, x_in.shape[1] * 2
    x_out = np.zeros((out_height, out_width))

    # Pad the input matrix to handle boundaries
    x_in_pad = np.zeros((x_in.shape[0] + 1, x_in.shape[1] + 1))
    x_in_pad[:x_in.shape[0], :x_in.shape[1]] = x_in
    x_in_pad[-1, :x_in.shape[1]] = x_in[-1, :]
    x_in_pad[:x_in.shape[0], -1] = x_in[:, -1]
    x_in_pad[-1, -1] = x_in[-1, -1]

    # Perform bilinear interpolation
    for i in range(out_height):
        for j in range(out_width):
            # Map the output coordinates to the corresponding coordinates in the padded input
            x = i / 2
            y = j / 2

            # Calculate the coordinates of the top left corner
            x1, y1 = int(np.floor(x)), int(np.floor(y))
            x2, y2 = x1 + 1, y1 + 1

            # Calculate interpolation weights
            wa = (x2 - x) * (y2 - y)
            wb = (x2 - x) * (y - y1)
            wc = (x - x1) * (y2 - y)
            wd = (x - x1) * (y - y1)

            # Compute the interpolated value
            x_out[i, j] = (wa * x_in_pad[x1, y1] + wb * x_in_pad[x1, y2] +
                           wc * x_in_pad[x2, y1] + wd * x_in_pad[x2, y2])

    return x_out
```
